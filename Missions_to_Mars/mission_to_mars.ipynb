````python
#import
import pandas as pd
from bs4 import BeautifulSoup as bs
from splinter import Browser
import requests
from webdriver_manager.chrome import ChromeDriverManager
````

````python
executable_path={'executable_path': ChromeDriverManager().install()}
browser=Browser('chrome', **executable_path, headless=False)
````

````python
#BS object by way of requests to URL
url="https://redplanetscience.com/"

browser.visit(url)
# response=requests.get(url)
html=browser.html

soup=bs(html, 'html.parser')
````

````python
soup
````

````python
# Scrape the [Mars News Site](https://redplanetscience.com/) and collect the 
    #latest News Title and Paragraph Text. Assign the text to variables that 
    # you can reference later.

slide_elem = soup.select_one('div', class_='list_text')
news_title = slide_elem.find('div', class_='content_title').get_text()
news_blurb = slide_elem.find('div', class_='article_teaser_body').get_text()
print(news_title)
print(news_blurb)
````

````python
# ### Mars Facts

# * Visit the Mars Facts webpage [here](https://galaxyfacts-mars.com) and 
# use Pandas to scrape the table containing facts about the planet including 
# Diameter, Mass, etc.

# * Use Pandas to convert the data to a HTML table string.
facts_url="https://galaxyfacts-mars.com"
tables=pd.read_html(facts_url)
tables
````

````python
#make a dataframe; drop the nonsense header; drop Earth column
df=tables[0]
new_header=df.iloc[0]
df=df[1:]
df.columns=new_header
df
````

````python
df=df.drop(['Earth'], axis=1)
df=df.reset_index(drop=True)
df=df.rename(columns={'Mars - Earth Comparison':" "})
````
